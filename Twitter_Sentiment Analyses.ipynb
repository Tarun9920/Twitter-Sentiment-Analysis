```python
"""
We will now import Textblob
we use 2 utility functions
1) cleaning function if a character other than alphanumeric comes it will change it to new one
2) classifier Function to find polarity of text after cleaning
"""
```

```python
from textblob import TextBlob
import re
```

```python
def clean_tweet(tweet):
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",tweet).split())
```

```python
def analyse_sentiment(tweet):
    analysis= TextBlob(clean_tweet(tweet))
    if analysis.sentiment.polarity>0:
        return 1
    elif analysis.sentiment.polarity==0:
        return 0
    else:
        return -1
```

```python
data['Sentiment']= np.array([analyse_sentiment(tweet) for tweet in data['Tweets']])
display(data.head(10))
```

```python
ps=0
ns=0
nn=0
for sa in data['Sentiment']:
    if sa>0.0:
        ps+=1
    elif sa<0.0:
        ns+=1
    else:
        nn+=1
```

```python
print("Percentage of positive sentiment : {}%".format(ps*100/len(data['Tweets'])))
print("Percentage of Negative Sentiment : {}%".format(ns*100/len(data['Tweets'])))
print("Percentage of Neutral sentiment : {}%".format(nn*100/len(data['Tweets'])))
```
